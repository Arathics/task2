```python
data2.drop(['PassengerId'],axis=1,inplace=True)
data2.columns
```

```python
# Convert relevant columns to numeric type if they are not already
for col in data2.columns:
    if data2[col].dtype == 'object':  # Check if the column is of object type (likely string)
        try:
            data2[col] = pd.to_numeric(data2[col], errors='coerce')  # Convert to numeric, replace non-convertibles with NaN
        except:
            print(f"Could not convert column {col} to numeric.")
```

```python
 Q1= data2.quantile(0.25)
 Q3= data2.quantile(0.75)
 IQR= Q3-Q1
 print(IQR)
```

```python
data2=data2[~((data2 < (Q1 - 1.5 * IQR)) |(data2 > (Q3 + 1.5 * IQR))).any(axis=1)]
data2
```
